Getting the Neighborhoods in Toronto using Foursquare API

```python
LIMIT=500
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
c=[0,0,0,0]
for i in range(len(aa)):
    
    df=toronto_df[toronto_df['Borough']==aa[i]]
    c[i] = getNearbyVenues(names=df['Neighborhood'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                                  )
```

Finding out the venues in each Neighborhood

```python
d=[0,0,0,0]
for i in range(len(c)):
    toronto_c_onehot = pd.get_dummies(c[i][['Venue Category']], prefix="", prefix_sep="")


    toronto_c_onehot['Neighborhood'] = c[i]['Neighborhood'] 


    fixed_columns = [toronto_c_onehot.columns[-1]] + list(toronto_c_onehot.columns[:-1])
    toronto_c_onehot = toronto_c_onehot[fixed_columns]

    d[i]=toronto_c_onehot
```

The mean of the frequency of occurrence of each category

```python
for i in range(len(c)):ci_grouped[i] = d[i].groupby('Neighborhood').mean().reset_index()
```

Each neighborhood along with the top 5 most common venues

```python
for i in range(len(ci_grouped)):
    for hood in ci_grouped[i]['Neighborhood']:
        print("----"+hood+"----")
        temp = ci_grouped[i][ci_grouped[i]['Neighborhood'] == hood].T.reset_index()
        temp.columns = ['venue','freq']
        temp = temp.iloc[1:]
        temp['freq'] = temp['freq'].astype(float)
        temp = temp.round({'freq': 2})
        print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head())
        print('\n')
```

The top 10 venues for each neighborhood.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
import numpy as np
```

```python
e=[0,0,0,0]
for i in range(len(ci_grouped)):
    num_top_venues = 10

    indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
    columns = ['Neighborhood']
    for ind in np.arange(num_top_venues):
        try:
            columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
        except:
            columns.append('{}th Most Common Venue'.format(ind+1))

    # create a new dataframe
    neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
    neighborhoods_venues_sorted['Neighborhood'] = ci_grouped[i]['Neighborhood']

    for ind in np.arange(ci_grouped[i].shape[0]):
        neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(ci_grouped[i].iloc[ind, :], num_top_venues)
    e[i]=neighborhoods_venues_sorted
e[0].head()
```

```python
from sklearn.cluster import KMeans
```

Running k-means to cluster the neighborhood into 5 clusters

```python
kmeans=[0,0,0,0]
for i in range(len(e)):
    
    kclusters = 5

    ci_grouped_clustering = ci_grouped[i].drop('Neighborhood', 1)

# run k-means clustering
    kmeans[i] = KMeans(n_clusters=kclusters, random_state=0).fit(ci_grouped_clustering)

    # check cluster labels generated for each row in the dataframe
kmeans[0].labels_[0:10] 
```

Merging the new dataframe that includes the cluster and the top 10 venues for each neighborhood with Neighborhood Dataframe

```python
toronto_merged=[0,0,0,0]
for i in range(len(e)):
    e[i].insert(0, 'Cluster Labels', kmeans[i].labels_)

    toronto_merged[i] = toronto_df[toronto_df['Borough']==aa[i]]

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
    toronto_merged[i] = toronto_merged[i].join(e[i].set_index('Neighborhood'), on='Neighborhood')

toronto_merged[0].head()
```
